```python
import plotly.graph_objects as go
import pandas as pd
from IPython.display import display, Image
```

```python
window = 50
```

```python
df_cart_pole_dqn_1 = pd.read_csv('results/cart_pole_DQN_rewards_1.csv')
df_cart_pole_dqn_2 = pd.read_csv('results/cart_pole_DQN_rewards_2.csv')
df_cart_pole_dqn_3 = pd.read_csv('results/cart_pole_DQN_rewards_3.csv')
df_cart_pole_dqn_4 = pd.read_csv('results/cart_pole_DQN_rewards_4.csv')
df_cart_pole_dqn_5 = pd.read_csv('results/cart_pole_DQN_rewards_5.csv')
```

```python
df_cart_pole_ddqn_1 = pd.read_csv('results/cart_pole_DDQN_rewards_1.csv')
df_cart_pole_ddqn_2 = pd.read_csv('results/cart_pole_DDQN_rewards_2.csv')
df_cart_pole_ddqn_3 = pd.read_csv('results/cart_pole_DDQN_rewards_3.csv')
df_cart_pole_ddqn_4 = pd.read_csv('results/cart_pole_DDQN_rewards_4.csv')
df_cart_pole_ddqn_5 = pd.read_csv('results/cart_pole_DDQN_rewards_5.csv')
```

```python
df_cart_pole_dqn = pd.DataFrame()
df_cart_pole_dqn['episode'] = df_cart_pole_dqn_1['episode']
df_cart_pole_dqn['reward_1'] = df_cart_pole_dqn_1['reward']
df_cart_pole_dqn['reward_2'] = df_cart_pole_dqn_2['reward']
df_cart_pole_dqn['reward_3'] = df_cart_pole_dqn_3['reward']
df_cart_pole_dqn['reward_4'] = df_cart_pole_dqn_4['reward']
df_cart_pole_dqn['reward_5'] = df_cart_pole_dqn_5['reward']
df_cart_pole_dqn['rolling_mean_1'] = df_cart_pole_dqn['reward_1'].rolling(window=window).mean()
df_cart_pole_dqn['rolling_mean_2'] = df_cart_pole_dqn['reward_2'].rolling(window=window).mean()
df_cart_pole_dqn['rolling_mean_3'] = df_cart_pole_dqn['reward_3'].rolling(window=window).mean()
df_cart_pole_dqn['rolling_mean_4'] = df_cart_pole_dqn['reward_4'].rolling(window=window).mean()
df_cart_pole_dqn['rolling_mean_5'] = df_cart_pole_dqn['reward_5'].rolling(window=window).mean()
df_cart_pole_dqn['rolling_mean_mean'] = df_cart_pole_dqn.iloc[:, 1:].mean(axis=1)
df_cart_pole_dqn['Var_min'] = df_cart_pole_dqn.iloc[50:, 1:-1].min(axis=1)
df_cart_pole_dqn['Var_max'] = df_cart_pole_dqn.iloc[50:, 1:-2].max(axis=1) 
df_cart_pole_dqn
```

```python
df_cart_pole_ddqn = pd.DataFrame()
df_cart_pole_ddqn['episode'] = df_cart_pole_ddqn_1['episode']
df_cart_pole_ddqn['reward_1'] = df_cart_pole_ddqn_1['reward']
df_cart_pole_ddqn['reward_2'] = df_cart_pole_ddqn_2['reward']
df_cart_pole_ddqn['reward_3'] = df_cart_pole_ddqn_3['reward']
df_cart_pole_ddqn['reward_4'] = df_cart_pole_ddqn_4['reward']
df_cart_pole_ddqn['reward_5'] = df_cart_pole_ddqn_5['reward']
df_cart_pole_ddqn['rolling_mean_1'] = df_cart_pole_ddqn['reward_1'].rolling(window=window).mean()
df_cart_pole_ddqn['rolling_mean_2'] = df_cart_pole_ddqn['reward_2'].rolling(window=window).mean()
df_cart_pole_ddqn['rolling_mean_3'] = df_cart_pole_ddqn['reward_3'].rolling(window=window).mean()
df_cart_pole_ddqn['rolling_mean_4'] = df_cart_pole_ddqn['reward_4'].rolling(window=window).mean()
df_cart_pole_ddqn['rolling_mean_5'] = df_cart_pole_ddqn['reward_5'].rolling(window=window).mean()
df_cart_pole_ddqn['rolling_mean_mean'] = df_cart_pole_ddqn[['rolling_mean_1', 'rolling_mean_2', 'rolling_mean_3', 'rolling_mean_4', 'rolling_mean_5']].mean(axis=1)
df_cart_pole_ddqn['Var_min'] = df_cart_pole_ddqn[['rolling_mean_1', 'rolling_mean_2', 'rolling_mean_3', 'rolling_mean_4', 'rolling_mean_5']].min(axis=1) 
df_cart_pole_ddqn['Var_max'] = df_cart_pole_ddqn[['rolling_mean_1', 'rolling_mean_2', 'rolling_mean_3', 'rolling_mean_4', 'rolling_mean_5']].max(axis=1)
df_cart_pole_ddqn
```

```python
gamma = 0.99
epsilon = 1.0 
epsilon_min = 0.01
epsilon_dec = 0.99
learning_rate = 0.001
episodes = 300
batch_size = 64

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df_cart_pole_dqn['episode'].to_list()[50:]+df_cart_pole_dqn['episode'].to_list()[50:][::-1],
    y=df_cart_pole_dqn['Var_max'].to_list()[50:]+df_cart_pole_dqn['Var_min'].to_list()[50:][::-1],
    fill='toself',
    fillcolor='rgba(0,100,80,0.2)',
    line_color='rgba(255,255,255,0)',
    showlegend=False,
    name='Sarsa',
))

fig.add_trace(go.Scatter(
    x=df_cart_pole_dqn['episode'][50:], y=df_cart_pole_dqn['rolling_mean_mean'][50:],
    line_color='rgb(0,100,80)',
    name='Cart Pole - DQN',
))

fig.add_trace(go.Scatter(
    x=df_cart_pole_ddqn['episode'].to_list()[50:]+df_cart_pole_ddqn['episode'].to_list()[50:][::-1],
    y=df_cart_pole_ddqn['Var_max'].to_list()[50:]+df_cart_pole_ddqn['Var_min'].to_list()[50:][::-1],
    fill='toself',
    fillcolor='rgba(244,196,48,0.2)',
    line_color='rgba(255,255,255,0)',
    showlegend=False,
    name='Q-Learning',
))

fig.add_trace(go.Scatter(
    x=df_cart_pole_ddqn['episode'][50:], y=df_cart_pole_ddqn['rolling_mean_mean'][50:],
    line_color='rgb(244,196,48)',
    name='Cart Pole - DDQN',
))

fig.add_annotation(x=1.25, y=0.8, xref='paper', yref='paper', 
                    text=f'<b>HiperparÃ¢metros:</b> <br>Î³= {gamma}, Îµ= {epsilon}<br>Îµ_min= {epsilon_min}, Îµ_dec= {epsilon_dec}<br> Î±= {learning_rate}, batch= {batch_size}', 
                    showarrow=False, font=dict(size=18, family="Arial",), align='center')

fig.update_traces(mode='lines')
fig.update_layout(title=f'DQN vs DDQN - Cart Pole 50 episodes moving average)', 
                  xaxis_title='Episodes', 
                  yaxis_title='Rewards', 
                  template='presentation', 
                  showlegend=True,
                  height=700,
                  width=1200,)
fig.show()
```
